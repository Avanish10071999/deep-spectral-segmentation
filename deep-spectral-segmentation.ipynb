```python
import os
import urllib.request
import tarfile

# Download URLs
voc_url = "http://host.robots.ox.ac.uk/pascal/VOC/voc2007/VOCtrainval_06-Nov-2007.tar"

# Paths
voc_tar_path = "/content/VOCtrainval_06-Nov-2007.tar"
voc_extract_path = "/content"

# Download Pascal VOC2007
if not os.path.exists(voc_tar_path):
    print("Downloading VOC2007...")
    urllib.request.urlretrieve(voc_url, voc_tar_path)

# Extract the tar file
print("Extracting...")
with tarfile.open(voc_tar_path) as tar:
    tar.extractall(path=voc_extract_path)

print("Done! VOC dataset is ready.")
```

```python
file_path = "/content/deep-spectral-segmentation-main/object-localization/datasets.py"

with open(file_path, "r") as f:
    content = f.read()

# Update the VOC dataset root path
patched_content = content.replace(
    'voc_root = "datasets/VOC2007"',
    'voc_root = "/content/VOCdevkit/VOC2007"'
)

with open(file_path, "w") as f:
    f.write(patched_content)

print("Patched datasets.py successfully.")
```

```python
file_path = "/content/deep-spectral-segmentation-main/object-localization/datasets.py"

# We'll overwrite the voc_root logic directly
with open(file_path, "r") as f:
    lines = f.readlines()

new_lines = []
for line in lines:
    if 'voc_root =' in line:
        new_lines.append('    voc_root = "/content/VOCdevkit/VOC2007"\n')
    else:
        new_lines.append(line)

with open(file_path, "w") as f:
    f.writelines(new_lines)

print("✅ Successfully forced VOC path to /content/VOCdevkit/VOC2007")
```
